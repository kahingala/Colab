```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2
```

```python
fig = plt.figure(figsize=(10, 7))   # Create a figure window
ax = fig.add_subplot(111, projection='3d')  # Add a 3D plot
```

```python
ax.plot_surface(X, Y, Z, cmap='viridis')
```

```python
ax.set_title("3D Plot of Z = X + Y")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
```

```python
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
```

```python
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
```

```python
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
```

```python
selected_features = ['hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']
```

```python
print(df.columns)
```

```python
selected_features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
df_stats = df[selected_features].dropna()
```

```python
pearson_corr = df_stats.corr(method='pearson')
spearman_corr = df_stats.corr(method='spearman')
```

```python
plt.figure(figsize=(14, 6))
```

```python
plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, square=True)
plt.title(" Pearson Correlation Heatmap")
```

```python
plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, square=True)
plt.title(" Spearman Rank Correlation Heatmap")
```

```python
plt.tight_layout()
plt.show()
```
